```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("data_3aug.csv", header = 1, parse_dates = ['Date/Time'], index_col = 'Date/Time')
df.head()
```

## Renaming Columns:

```python
df.columns = ['air_temp_avg', 'air_temp_max', 'air_temp_min', 'dew_avg',
       'dew_min', 'solar_rad_avg', 'vpd_avg', 'vpd_min', 'rel_humi_avg',
       'rel_humi_max', 'rel_humi_min', 'precp_sum', 'leaf_wetness_time(min)',
       'wind_speed_avg', 'wind_speed_max', 'wind_gust_max', 'wind_dir_last',
       'eag_sm1_avg', 'eag_sm2_avg', 'eag_sm3_avg', 'eag_sm4_avg',
       'eag_sm5_avg', 'eag_sm6_avg', 'VIC1_avg', 'VIC2_avg', 'VIC3_avg',
       'VIC4_avg', 'VIC5_avg', 'VIC6_avg', 'soil_temp1_avg', 'soil_temp1_max',
       'soil_temp1_min', 'soil_temp2_avg', 'soil_temp2_max', 'soil_temp2_min',
       'soil_temp3_avg', 'soil_temp3_max', 'soil_temp3_min', 'soil_temp4_avg',
       'soil_temp4_max', 'soil_temp4_min', 'soil_temp5_avg', 'soil_temp5_max',
       'soil_temp5_min', 'soil_temp6_avg', 'soil_temp6_max', 'soil_temp6_min',
       'sp_last', 'bat_last', 'del_avg', 'del_max', 'del_min', 'sun_time',
       'Daily ET0 [mm]']
```

## Reversing DataFrame:

```python
df = df[::-1]
df.head()
```

## Dealing with Null Values:

```python
df['Daily ET0 [mm]'].fillna(0, inplace = True)
df.dropna(inplace = True)
```

## Selecting Only Significant Features:

```python
sign = ['soil_temp1_avg','soil_temp2_avg','soil_temp3_avg','eag_sm1_avg', 'eag_sm2_avg', 'eag_sm3_avg', 'precp_sum']
df1 = df[sign]
df1.reset_index(inplace = True)
df1.head()
```

## Selecting Records such that all the previous 5 zero records of precipitation sum would be selected wherever first non-zero precipitation record occurs.
That way we would be able to determine the average value of respective feature and we get the idea what value should be set as a threshold.

```python
val = []
for i, data in df1.iterrows():
  if data['precp_sum'] != 0:
    temp = df1.iloc[i-5:i].values
    val.append(temp)
```

## Combining Above Result into Single Array:

```python
comb_val = np.vstack(val)
```

## Generating Final Dataframe:

```python
final_df = pd.DataFrame(comb_val, columns = ['Date/Time', 'soil_temp1_avg', 'soil_temp2_avg', 'soil_temp3_avg', 'eag_sm1_avg','eag_sm2_avg','eag_sm3_avg', 'precp_sum'])
final_df.drop_duplicates()
final_df.drop('Date/Time', inplace = True, axis = 1)
```

```python
final_df.head()
```

## Changing Datatype of Each Column in final_df:

```python
for c in final_df.columns:
  final_df[c] = final_df[c].astype('float32')
```

## Generating Result Dataframe:

```python
col = ['soil_temp1_avg', 'soil_temp2_avg', 'soil_temp3_avg', 'eag_sm1_avg','eag_sm2_avg', 'eag_sm3_avg']
vals = final_df[col].describe().iloc[1,:].values
std = final_df[col].describe().iloc[2,:].values
```

```python
val_with_max_error = vals + std
```

```python
res = pd.DataFrame({'Values': val_with_max_error}, index=['SoilTemp1 >', 'SoilTemp2 >', 'SoilTemp3 >', 'Moisture1 >', 'Moisture2 >', 'Moisture3 >'])
res['Values'] = round(res['Values'],1)
print(res)
```
